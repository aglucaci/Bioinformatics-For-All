```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from scipy import stats
import scikit_posthocs as sp

# Generate Data for Line Graph
np.random.seed(42)
x = np.linspace(0, 10, 100)
y1 = np.sin(x) + np.random.normal(0, 0.1, len(x))
y2 = np.cos(x) + np.random.normal(0, 0.1, len(x))
y3 = np.sin(2*x) + np.random.normal(0, 0.1, len(x))

# Create Line Plot with Multiple Lines
plt.figure(figsize=(6,4), dpi=300)
plt.plot(x, y1, label="Sine Wave", linewidth=2, color='darkblue')
plt.plot(x, y2, label="Cosine Wave", linewidth=2, color='darkred')
plt.plot(x, y3, label="Double Frequency Sine", linewidth=2, color='darkgreen')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Multiple Line Plot")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Generate Data for Box Plot
group_labels = ['A', 'B', 'C', 'D']
data = [np.random.normal(loc, 1, 20) for loc in range(4)]

# Convert to DataFrame for Seaborn
df = pd.DataFrame({
    "Group": np.repeat(group_labels, 20),
    "Value": np.concatenate(data)
})

# Perform Pairwise Statistical Tests (ANOVA + t-tests)
anova_stat, anova_p = stats.f_oneway(*data)

# Compute pairwise t-tests
pairwise_pvalues = {}
for i in range(len(group_labels)):
    for j in range(i + 1, len(group_labels)):
        t_stat, p_val = stats.ttest_ind(data[i], data[j])
        pairwise_pvalues[(i, j)] = p_val

# Plot Box Plot with Statistical Significance
plt.figure(figsize=(6,4), dpi=300)
sns.boxplot(x="Group", y="Value", data=df, palette="coolwarm", linewidth=2)
plt.title("Box Plot with Pairwise Statistical Test")

# Annotate Statistical Significance
y_max = df["Value"].max() + 0.5

for (i, j), p_val in pairwise_pvalues.items():
    if p_val < 0.05:
        plt.plot([i, i, j, j], [y_max, y_max + 0.1, y_max + 0.1, y_max], color='black')
        plt.text((i+j)/2, y_max + 0.15, f"p={p_val:.3f}", ha='center', fontsize=10)
        y_max += 0.3

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
