```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
subcategories = ['Subcat A', 'Subcat B', 'Subcat C']
data = np.array([
    [5, 3, 4],
    [7, 2, 5],
    [6, 4, 3],
    [4, 5, 2]
])

# Colors for the subcategories
colors = ['#4CAF50', '#2196F3', '#FFC107']

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 7))

# Plot stacked bars
bottom = np.zeros(len(categories))
for i, color in enumerate(colors):
    ax.bar(categories, data[:, i], bottom=bottom, color=color, label=subcategories[i])
    bottom += data[:, i]

# Add labels, title, and legend
ax.set_xlabel('Categories', fontsize=14)
ax.set_ylabel('Values', fontsize=14)
ax.set_title('Stacked Bar Chart Example', fontsize=16)
ax.legend(title='Subcategories', fontsize=12, title_fontsize=13)

# Customize gridlines
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.set_axisbelow(True)

# Remove top and right spines for a cleaner look
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Customize ticks
ax.tick_params(axis='both', which='major', labelsize=12)

# Adjust layout for better spacing
plt.tight_layout()

# Save the figure as a high-quality PNG file
plt.savefig('stacked_bar_chart.png', dpi=300)

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
subcategories = ['Subcat A', 'Subcat B', 'Subcat C']
data = np.array([
    [5, 3, 4],
    [7, 2, 5],
    [6, 4, 3],
    [4, 5, 2]
])

# Create a DataFrame for better handling in seaborn
df = pd.DataFrame(data, columns=subcategories, index=categories)
df['Category'] = df.index

# Melt the DataFrame for seaborn
df_melted = df.melt(id_vars='Category', var_name='Subcategory', value_name='Value')

# Set the style and palette
sns.set(style="whitegrid")
palette = sns.color_palette("Set2")

# Create the plot
plt.figure(figsize=(10, 7))
bar_plot = sns.barplot(
    data=df_melted,
    y='Category',
    x='Value',
    hue='Subcategory',
    palette=palette,
    ci=None
)

# Add labels, title, and legend
bar_plot.set_ylabel('Categories', fontsize=14)
bar_plot.set_xlabel('Values', fontsize=14)
bar_plot.set_title('Stacked Horizontal Bar Chart Example', fontsize=16)
bar_plot.legend(title='Subcategories', fontsize=12, title_fontsize=13)

# Customize ticks
bar_plot.tick_params(axis='both', which='major', labelsize=12)

# Adjust layout for better spacing
plt.tight_layout()

# Save the figure as a high-quality PNG file
plt.savefig('stacked_horizontal_bar_chart_seaborn.png', dpi=300)

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
subcategories = ['Subcat A', 'Subcat B', 'Subcat C']
data = np.array([
    [5, 3, 4],
    [7, 2, 5],
    [6, 4, 3],
    [4, 5, 2]
])

# Sample standard errors for each subcategory in each category
errors = np.array([
    [0.5, 0.3, 0.4],
    [0.7, 0.2, 0.5],
    [0.6, 0.4, 0.3],
    [0.4, 0.5, 0.2]
])

# Create a DataFrame for better handling in seaborn
df = pd.DataFrame(data, columns=subcategories, index=categories)
df_errors = pd.DataFrame(errors, columns=subcategories, index=categories)
df['Category'] = df.index
df_errors['Category'] = df.index

# Melt the DataFrame for seaborn
df_melted = df.melt(id_vars='Category', var_name='Subcategory', value_name='Value')
df_errors_melted = df_errors.melt(id_vars='Category', var_name='Subcategory', value_name='Error')

# Merge the values and errors dataframes
df_combined = pd.merge(df_melted, df_errors_melted, on=['Category', 'Subcategory'])

# Set the style and palette
sns.set(style="whitegrid")
palette = sns.color_palette("Set2")

# Create the plot
plt.figure(figsize=(10, 7))
bar_plot = sns.barplot(
    data=df_combined,
    y='Category',
    x='Value',
    hue='Subcategory',
    palette=palette,
    errorbar=None
)

# Add error bars manually
for index, row in df_combined.iterrows():
    plt.errorbar(
        x=row['Value'],
        y=row['Category'],
        xerr=row['Error'],
        fmt='none',
        c='black',
        capsize=5
    )

# Add labels, title, and legend
bar_plot.set_ylabel('Categories', fontsize=14)
bar_plot.set_xlabel('Values', fontsize=14)
bar_plot.set_title('Stacked Horizontal Bar Chart with Error Bars', fontsize=16)
bar_plot.legend(title='Subcategories', fontsize=12, title_fontsize=13)

# Customize ticks
bar_plot.tick_params(axis='both', which='major', labelsize=12)

# Adjust layout for better spacing
plt.tight_layout()

# Save the figure as a high-quality PNG file
plt.savefig('stacked_horizontal_bar_chart_seaborn_errorbars.png', dpi=300)

# Display the chart
plt.show()
```
