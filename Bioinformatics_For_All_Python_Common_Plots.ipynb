t-Test: A statistical test used to compare the means of two groups to determine if they are significantly different from each other.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats

# Sample data
np.random.seed(0)
group1 = np.random.normal(10, 2, 100)
group2 = np.random.normal(12, 2, 100)
data = pd.DataFrame({'Group': ['Group 1']*100 + ['Group 2']*100, 'Value': np.concatenate([group1, group2])})

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Value', data=data, palette="Set2")
plt.title('Box Plot with t-Test', fontsize=16)

# Perform t-test
t_stat, p_val = stats.ttest_ind(group1, group2)
plt.text(0.5, 18, f't-statistic: {t_stat:.2f}\np-value: {p_val:.2e}', ha='center', va='top', fontsize=12, color='red')

plt.show()
```

Kruskal-Wallis Test: A test used to compare more than two groups that are independent, and assesses whether there is a statistically significant difference in their distributions.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy.stats import kruskal

# Sample data
np.random.seed(0)
group1 = np.random.normal(10, 2, 100)
group2 = np.random.normal(12, 2, 100)
group3 = np.random.normal(11, 2, 100)
data = pd.DataFrame({'Group': ['Group 1']*100 + ['Group 2']*100 + ['Group 3']*100, 'Value': np.concatenate([group1, group2, group3])})

# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Group', y='Value', data=data, palette="pastel")
plt.title('Violin Plot with Kruskal-Wallis Test', fontsize=16)

# Perform Kruskal-Wallis test
h_stat, p_val = kruskal(group1, group2, group3)
plt.text(1, 15, f'H-statistic: {h_stat:.2f}\np-value: {p_val:.2e}', ha='center', va='top', fontsize=12, color='red')

plt.show()
```

```python
import pandas as pd
import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, Label
from bokeh.transform import dodge
from scipy.stats import ttest_rel

# Enable output to Jupyter Notebook
output_notebook()

# Sample data
np.random.seed(0)
before = np.random.normal(10, 2, 100)
after = np.random.normal(12, 2, 100)
data = pd.DataFrame({'Before': before, 'After': after})

# Perform paired t-test
t_stat, p_val = ttest_rel(before, after)

# Prepare data for plotting
categories = ['Before', 'After']
values = [data['Before'].mean(), data['After'].mean()]
errors = [data['Before'].std(), data['After'].std()]

source = ColumnDataSource(data=dict(categories=categories, values=values, errors=errors))

# Create the bar chart
p = figure(x_range=categories, title="Paired Bar Chart with t-Test", y_axis_label='Value',)
p.vbar(x=dodge('categories', 0.0, range=p.x_range), top='values', width=0.4, source=source, legend_field="categories",
       line_color='black', alpha=0.5)

# Add error bars
p.add_layout(Label(x=0, y=max(values) + max(errors) + 1, text=f't-statistic: {t_stat:.2f}\np-value: {p_val:.2e}', text_color='red'))

# Add labels to the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show plot
show(p)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Sample data
np.random.seed(0)
group1 = np.random.normal(10, 2, 100)
group2 = np.random.normal(12, 2, 100)
group3 = np.random.normal(11, 2, 100)
data = pd.DataFrame({'Group': ['Group 1']*100 + ['Group 2']*100 + ['Group 3']*100, 'Value': np.concatenate([group1, group2, group3])})

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Value', data=data, palette="Set2")
plt.title('Boxplot with Pairwise T-tests')

# Perform pairwise t-tests
comparisons = [('Group 1', 'Group 2'), ('Group 1', 'Group 3'), ('Group 2', 'Group 3')]
p_values = []
for comp in comparisons:
    group1_data = data[data['Group'] == comp[0]]['Value']
    group2_data = data[data['Group'] == comp[1]]['Value']
    t_stat, p_val = ttest_ind(group1_data, group2_data)
    p_values.append((comp[0], comp[1], p_val))

# Annotate the plot with p-values
y_max = data['Value'].max()
for i, (group1, group2, p_val) in enumerate(p_values):
    x1, x2 = i, i + 1
    y, h = y_max + (i + 1) * 0.5, 0.2
    plt.plot([x1, x1, x2, x2], [y, y + h, y + h, y], lw=1.5, color='k')
    plt.text((x1 + x2) * .5, y + h, f"p = {p_val:.3e}", ha='center', va='bottom', color='k')

plt.show()
```

```python
import numpy as np
from scipy.linalg import expm

# Define the genetic code and codons
genetic_code = {
    'AAA': 'K', 'AAC': 'N', 'AAG': 'K', 'AAU': 'N',
    'ACA': 'T', 'ACC': 'T', 'ACG': 'T', 'ACU': 'T',
    'AGA': 'R', 'AGC': 'S', 'AGG': 'R', 'AGU': 'S',
    'AUA': 'I', 'AUC': 'I', 'AUG': 'M', 'AUU': 'I',
    'CAA': 'Q', 'CAC': 'H', 'CAG': 'Q', 'CAU': 'H',
    'CCA': 'P', 'CCC': 'P', 'CCG': 'P', 'CCU': 'P',
    'CGA': 'R', 'CGC': 'R', 'CGG': 'R', 'CGU': 'R',
    'CUA': 'L', 'CUC': 'L', 'CUG': 'L', 'CUU': 'L',
    'GAA': 'E', 'GAC': 'D', 'GAG': 'E', 'GAU': 'D',
    'GCA': 'A', 'GCC': 'A', 'GCG': 'A', 'GCU': 'A',
    'GGA': 'G', 'GGC': 'G', 'GGG': 'G', 'GGU': 'G',
    'GUA': 'V', 'GUC': 'V', 'GUG': 'V', 'GUU': 'V',
    'UAA': '*', 'UAC': 'Y', 'UAG': '*', 'UAU': 'Y',
    'UCA': 'S', 'UCC': 'S', 'UCG': 'S', 'UCU': 'S',
    'UGA': '*', 'UGC': 'C', 'UGG': 'W', 'UGU': 'C',
    'UUA': 'L', 'UUC': 'F', 'UUG': 'L', 'UUU': 'F'
}

codons = list(genetic_code.keys())

# Define synonymous and nonsynonymous changes
def is_synonymous(codon1, codon2):
    return genetic_code[codon1] == genetic_code[codon2]

# Compute the substitution rate matrix
def compute_rate_matrix(codons, omega, kappa):
    n = len(codons)
    Q = np.zeros((n, n))

    for i, codon1 in enumerate(codons):
        for j, codon2 in enumerate(codons):
            if i != j:
                diff = sum(1 for a, b in zip(codon1, codon2) if a != b)
                if diff == 1:
                    if is_synonymous(codon1, codon2):
                        Q[i, j] = 1  # Synonymous rate
                    else:
                        Q[i, j] = omega  # Nonsynonymous rate

    # Normalize and apply transition/transversion ratio (kappa)
    for i in range(n):
        for j in range(n):
            if Q[i, j] != 0:
                base1 = codons[i][[a != b for a, b in zip(codons[i], codons[j])].index(True)]
                base2 = codons[j][[a != b for a, b in zip(codons[i], codons[j])].index(True)]
                if base1 in 'AG' and base2 in 'AG' or base1 in 'CTU' and base2 in 'CTU':
                    Q[i, j] *= kappa
        Q[i, i] = -np.sum(Q[i, :])

    return Q

# Set model parameters
omega = 0.1  # Nonsynonymous to synonymous rate ratio
kappa = 2.0  # Transition/transversion ratio

# Compute the rate matrix
Q = compute_rate_matrix(codons, omega, kappa)

# Compute the transition probability matrix
t = 1.0  # Time
P = expm(Q * t)

# Print results
print("Rate Matrix (Q):")
print(Q)
print("\nTransition Probability Matrix (P):")
print(P)
```

```python
import plotly.graph_objects as go
import numpy as np

# Generate node labels dynamically
num_sources = 10
num_middle = 5
num_targets = 10

labels = [f"Source {i+1}" for i in range(num_sources)] + \
         [f"Middle {i+1}" for i in range(num_middle)] + \
         [f"Target {i+1}" for i in range(num_targets)]

# Assign base colors to nodes (RGB tuples)
np.random.seed(42)
node_colors = {label: (np.random.randint(50, 255), np.random.randint(50, 255), np.random.randint(50, 255))
               for label in labels}

# Generate random links between sources, middle, and targets
source_nodes = np.random.choice(range(num_sources), size=15, replace=True).tolist()
target_nodes = np.random.choice(range(num_sources, num_sources + num_middle), size=15, replace=True).tolist()
middle_to_target = np.random.choice(range(num_sources, num_sources + num_middle), size=15, replace=True).tolist()
final_targets = np.random.choice(range(num_sources + num_middle, len(labels)), size=15, replace=True).tolist()

# Flow values (randomized for variation)
values_1 = np.random.randint(5, 50, size=15).tolist()
values_2 = np.random.randint(5, 50, size=15).tolist()

# Combine links
sources = source_nodes + middle_to_target
targets = target_nodes + final_targets
values = values_1 + values_2

# Duplicate each link for two colors per edge
extended_sources = []
extended_targets = []
extended_values = []
extended_colors = []

for i in range(len(sources)):
    src_label = labels[sources[i]]
    tgt_label = labels[targets[i]]

    src_r, src_g, src_b = node_colors[src_label]
    tgt_r, tgt_g, tgt_b = node_colors[tgt_label]

    # Source-colored half of the link
    extended_sources.append(sources[i])
    extended_targets.append(targets[i])
    extended_values.append(values[i] * 0.5)  # Half value
    extended_colors.append(f"rgba({src_r}, {src_g}, {src_b}, 0.9)")

    # Target-colored half of the link
    extended_sources.append(sources[i])
    extended_targets.append(targets[i])
    extended_values.append(values[i] * 0.5)  # Half value
    extended_colors.append(f"rgba({tgt_r}, {tgt_g}, {tgt_b}, 0.9)")

# Create Sankey diagram
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels,
        color=[f"rgba({r},{g},{b},1.0)" for r, g, b in node_colors.values()],  # Node colors
    ),
    link=dict(
        source=extended_sources,
        target=extended_targets,
        value=extended_values,
        color=extended_colors,  # Two-color effect for each edge
    ),
))

# Set layout and show the figure
fig.update_layout(title_text="Sankey Diagram with Two-Color Edge Effect", font_size=12)
fig.show()
```
